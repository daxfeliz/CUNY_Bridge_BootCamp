# What is [VSCode](https://code.visualstudio.com/)?

Visual Studio Code, also commonly referred to as VS Code, is a source-code editor made by Microsoft with the Electron Framework, for Windows, Linux and macOS. Features include support for debugging, syntax highlighting, intelligent code completion, snippets, code refactoring, and embedded Git. VSCode is a free, open source software that is designed to make programming easier! 

#### Note: I'm very old-school and use vim primarily, however a lot of modern astronomers use VScode! With that said, I'm fairly new to VScode myself!

## Getting Started:

First, please visit the [__download page__](https://code.visualstudio.com/Download) to begin the download of the VSCode software. It should be a `.zip` file. Use your __Finder__ to click on this `.zip` file (it's likely in your __`Downloads`__ folder) and it will unzip an _Application_ called ___Visual Studios Code___.

Next, drag this _Application_ file into your ___Applications___ folder. Then click on it to start it up. You should see a Welcome Screen like this:


<img src="./images/vscode1.png" width="60%">

If you'd like to, select your theme. Personally, I prefer "_Dark Modern_"

## Next, since we already have Anaconda installed, we'll need to make our  it's properly recognized by VSCode.
To do so, first we need to install the Python extension.

Pull up the "__Command Palette__" by typing `command + shift + P` or going to the toolbar at the top and selecting __View__ then then __Command Palette__. It should pull up a text box like this:

<img src="./images/vscode2.png" width="60%">

In the text box, enter `Extensions: Install Extensions`. It should then pull up a window like this:

<img src="./images/vscode3.png" width="60%">

We're going to want to install the `Python` extension. However, this Python will not be the same version of Python as our Anaconda installation.

Anaconda installs Python in a default directory located at:

```
/Users/YOURUSERNAME/opt/anaconda3/bin/python
```

VSCode will try to use the Python extension installation which is located at:
```
/usr/bin/python3 
```

To make sure VSCode is using Anaconda, we now need to go to the toolbar at the top, select __Code__, then __Preferences__, then __Settings__. It should pull up a window that looks like this:

<img src="./images/vscode4.png" width="60%">

From here, in the main window (under the __Settings__ tab), click on __>Extensions__, then scroll down to __Python__ and click on it. You should then see a window that looks like this:

<img src="./images/vscode5.png" width="60%">

We then want to click on the blue text that says ___Edit in settings.json___:
Which will pull up a window that looks like this:

<img src="./images/vscode6a.png" width="60%">

We want to change the line that says:

__`"python.defaultInterpreterPath": /usr/bin/python3`__  

to read as 

__`"python.defaultInterpreterPath": /Users/YOURUSERNAME/opt/anaconda3/bin/python`__  

#### _replace YOURUSERNAME with your macbook user name!_

So this window should now look like:

<img src="./images/vscode6b.png" width="60%">

# Creating a Python script with VSCode
Now that we're properly configured, let's close our tabs for __Extensions__ and __Settings__ and then go to the toolbar at the top and select __File__, then __Open__. This will prompt us to select a folder to initialize our _Workspace_ from. Let's select __"Notebook_03__" from our copied CUNY Bridge Bootcamp folders on our Desktops.

Once you do, you should see a new toolbar on the left has appeared, like this:


<img src="./images/vscode7.png" width="60%">

To create a blank Python file, in the main window under "__Start__", click on __New File__. You will be prompted to select an _Interpreter_ for this file. Select `_Python_`.

A new tab in our workspace will appear, titled: `'Untitled-1'`. As you can see in the lower right, our base environment is with `conda`:

<img src="./images/vscode8.png" width="60%">

Let's type `command + s` to save this file and let's called `"hello.py"`.

By using the `.py` file extension, you tell VS Code to interpret this file as a Python program, so that it evaluates the contents with the Python extension and the selected interpreter.


Let's now, modify the contents of our hello.py file in the main window:

type:

```python
msg = "Roll a dice"
print(msg)
```

When you start typing print, notice how VScode presents auto-completion options. 
<img src="./images/vscode9.png" width="60%">
This is super useful! Press `tab` to auto-complete commands!

Auto-completions work for standard Python modules as well as other packages you've installed into the environment of the selected Python interpreter. It also provides completions for methods available on object types. For example, because the msg variable contains a `"string"`, VScode provides string methods when you type msg.:
<img src="./images/vscode10.png" width="60%">


#### Notice, I mispelled "print" and it is underlined with a yellow sqiggly line. This is super useful for catching typos!

## Configure and run the debugger
If you highlight your cursor next to the line numbers, you'll see a red dot appear on the line number you hover over. This is called a `__breakpoint__`. Let's place a breakpoint on Line 2 (`print(msg)`).

<img src="./images/vscode13.png" width="60%">

Next, on the left-most tool bar, there should be an icon that looks like a bug on a Play button. This is the __`Run & Debug`__ button. You can also pull up this window with `command + shift + d`. By clicking this button and  clicking the new blue `Run and Debug` button, a text box should appear like this:

<img src="./images/vscode14.png" width="60%">

Select `Python File`, which is the configuration that runs the current file shown in the editor using the currently selected Python interpreter.

Start the debugger by clicking on the down-arrow next to the run button on the editor, and selecting __Debug Python File__ in Terminal.

<img src="./images/vscode15.png" width="60%">

The debugger will stop at the first line of the file breakpoint. The current line is indicated with a yellow arrow in the left margin. If you examine the Local variables window at this point, you will see now defined msg variable appears in the Local panel:

<img src="./images/vscode16.png" width="60%">

A debug toolbar appears in a new window on the lower left called a __"Call Stack"__ and when hovering your cursor over the debug session, you'll see a new tool bar with the following commands from left to right: continue (`F5`), step over (`F10`), step into (`F11`), step out (`shift + F11`), restart (`command + shift + F5`), and stop (`shift+F5`):
<img src="./images/vscode17.png" width="60%">

Let's build upon the previous example by using packages.

In Python, packages are how you obtain any number of useful code libraries, typically from PyPI, that provide additional functionality to your program. For this example, you use the `numpy` package to generate a random number.

In the main window where our `hello.py` script is located, let's add the lines:
```python
import numpy as np

msg = "Roll a dice"
print(msg)

print(np.random.randint(1,6))
```

If we then remove our _breakpoint_ and then click __Debug Python File__ from the pull down menu on the Play button in the upper right tool bar, we'll see that our code fully compiles without error and generates a random dice roll:

<img src="./images/vscode18.png" width="60%">
